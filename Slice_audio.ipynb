```python
import librosa
import soundfile as sf
from tqdm import tqdm
import numpy as np
import os
from scipy.io import wavfile
```

```python
directory = "C:/Users/ASUS/OneDrive - BUET/Desktop/SR_DSP/DataSet/Primary Data"
slice_length = 3  # in seconds
for root, subdirectories, files in os.walk(directory):
    for subdirectory in subdirectories:
            for filename in os.listdir(os.path.join(root, subdirectory)):
                x,sr = librosa.load(os.path.join(root, subdirectory)+"/"+ str(filename),duration = 300,sr = 44100)
                x = np.array(x)
                x.resize(300*sr)
                slices = np.arange(0, len(x)/sr+1, slice_length, dtype=np.int)
                c=1
                for start, end in zip(slices[:-1], slices[1:]):
                       
                        start_audio = start * sr
                        end_audio = end * sr
                        audio_slice = x[start_audio: end_audio]
                        sf.write(os.path.join(root, subdirectory)+"/"+str(subdirectory)+"_"+str(c)+".wav", audio_slice, sr, 'PCM_16')
                        c = c + 1
                        
                
print("Ended data slicing....")
```
