```python
import librosa
import numpy as np
import soundfile as sf
from tqdm import tqdm
import os
    
directory = "C:/Users/ASUS/OneDrive - BUET/Desktop/SR_DSP/Recordings for project/Data/ID_36"
c =0
for filename in os.listdir(directory):
    c=c+1
    x, sr = librosa.load(directory+"/"+str(filename),duration=1,sr=44100)
    x = np.array(x)
    
    x.resize(44100)
    sf.write("C:/Users/ASUS/OneDrive - BUET/Desktop/SR_DSP/Recordings for project/Data/ID_36/"+"ID_36_"+str(c)+".wav", x, sr, 'PCM_16')
    print(x.shape)
```
